# 定义

- **supervised problem** : A problem in which we are required to predict a value is known
> 监督学习问题：一个问题，这个问题是我们需要预测一个值，这个值是已知的
- **unsupervised problem** : Unsupervised datasets do not have a target associated with them and in general, are more challenging to deal with when compared to supervised problems.
> 无监督学习问题：无监督数据集没有与之相关的目标，一般来说，与监督问题相比，无监督问题更具挑战性。
- **Clustering** : Clustering is the task of grouping together a set of objects in a way that objects in the same cluster are more similar to each other than to objects in other clusters.
> 聚类：聚类是将一组对象分组的任务，使得同一聚类中的对象彼此之间的相似性大于其他聚类中的对象。
- **PCA**: Principal Component Analysis(PCA) is a dimensionality reduction technique that projects the data into a lower dimensional space.
> 主成分分析(PCA)是一种降维技术，它将数据投影到较低维空间中。
- **t-SNE**: t-distributed Stochastic Neighbor Embedding(t-SNE), is a non-linear dimensionality reduction technique that is particularly well-suited for embedding high-dimensional data into a space of two or three dimensions, which can then be visualized in a scatter plot.
> t分布随机邻域嵌入(t-SNE)是一种非线性降维技术，特别适用于将高维数据嵌入到二维或三维空间中，然后在散点图中可视化。

# Supervised and Unsupervised Learning

## Part1



- As we see in figure 1,every row of the data is associated with a target or label.
> 如图1所示，数据的每一行都与一个目标或标签相关联。
- The columns are different features and rows represent different data points which are usually called samples.
> 列是不同的特征，行代表不同的数据点，通常称为样本。
- The example shows ten samples with ten features and a target variable which can be either a number or a category.
> 该示例显示了具有十个特征和一个目标变量的十个样本，该目标变量可以是数字或类别。
- If the target is categorical, the problem becomes a classification problem.
> 如果目标是分类的，则问题变成了分类问题。
- And if the target is a real number, the problem is defined as a regression problem.
> 如果目标是实数，则问题被定义为回归问题。
- Thus,supervised problems can be divided into two sub-classes:
    - classification: predicting a category, e.g. dog or cat
    - Regression: predicting a value, e.g. house proces.
> 因此，监督问题可以分为两个子类：
    - 分类：预测类别，例如狗或猫
    - 回归：预测值，例如房屋价格。
- Another type of machine learning peoblem is the unsupervised type.
> 另一种类型的机器学习问题是无监督类型。
- Unsupervised datasets do not have a target associated with them and in general,are more challenging to deal with when compared to supervised problems.
> 无监督数据集没有与之相关的目标，一般来说，与监督问题相比，无监督问题更具挑战性。
- Let's say you work in a financial firm which deals with credit card transactions. There is a lot of data that comes in every second.
> 假设您在一家处理信用卡交易的金融公司工作。 每秒都会有大量数据进入。
- The only problem is that it is difficult to find humans who will mark each and every transaction either as a vaild or genuine transaction or a fraud.
> 唯一的问题是很难找到人类，他们会将每笔交易标记为有效或真实交易或欺诈。
- When we do not have any information about a transaction being fraud or genuine, the problem becomes an unsupervised problem.
> 当我们没有任何关于交易是欺诈还是真实的信息时，问题就变成了无监督的问题。
- To tackle these kinds of problems we have to think about how many clusters can data be divided into.
> 要解决这些问题，我们必须考虑数据可以分成多少个集群。
- For a fraud detection problem,we can say that data can be divided into two classes(fraud or genuine).
> 对于欺诈检测问题，我们可以说数据可以分为两类（欺诈或真实）。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets
from sklearn import manifold

%matplotlib inline
data = datasets.fetch_openml(
    'mnist_784',
    version=1,
    return_X_y=True
)

pixel_values, targets = data
single_image = pixel_values.iloc[1, :].values.reshape(28, 28)
plt.imshow(single_image, cmap='gray')
```

> 此步骤创建数据的t-SNE转换。 我们仅使用两个组件，因为我们可以在二维设置中很好地可视化它们。

> 在这种情况下，transformed_data是一个形状为3000 * 2（3000行和2列）的数组。
