# 【W1-4】Train the model with gradient descent


## 1.

Gradient descent is an algorithm for finding values of parameters w and b that minimize the cost function J. 
> 梯度下降是一种用于找到最小化成本函数J的参数w和b的算法。
repeat until convergence {
> 重复直到收敛
$$
w = w - \alpha \frac{\partial J(w,b)}{\partial w}
$$
$$
b = b - \alpha \frac{\partial J(w,b)}{\partial b}
$$
}
When $\frac{\partial J(w,b)}{\partial w}$ is a negative number(less than zero), what happens to $w$ after one update step?
> 当$\frac{\partial J(w,b)}{\partial w}$是一个负数(小于零)时，经过一次更新步骤后，$w$会发生什么？

- [ ] A. $w$ decreases
- [ ] B. It is not possible to tell if $w$ will increase or decrease.
- [ ] C. $w$ increases
- [ ] D. $w$ stays the same

<div class="alert alert-success">
<b>Currect :C</b>
<p>
The learning rare is always a positive number,so if you take W minus a negative number, you end up with a new value of W that is larger(more positive).
</p>
<hr>
<i>
学习率始终是一个正数，因此如果您将W减去一个负数，您最终会得到一个更大(更正数)的新值W。
</i>
</div>
