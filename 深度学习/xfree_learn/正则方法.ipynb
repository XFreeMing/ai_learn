# Reguralization
is the process of adding information in order to solve an illposed problem or to prevent overfitting
> 是解决不适定问题或防止过拟合的过程
- illposed problem
> 是一个微小的输入变化可能导致输出结果巨大变化的问题

## 1. L1 and L2 Regularization
- L1 Regularization
> L1正则化是指权重的绝对值之和

$$
Cost(w) = J(w) + \lambda \sum_{i=1}^{n} |w_i|
$$

- L2 Regularization
> L2正则化是指权重的平方和

$$
Cost(w) = J(w) + \lambda \sum_{i=1}^{n} w_i^2
$$


## 2. Dropout
> Dropout是指在训练过程中随机删除一些神经元，以减少过拟合


## 3. Data Augmentation
> 数据增强是指通过对原始数据进行一些变换，以增加数据量，减少过拟合


## 4. Early Stopping
> 提前停止是指在训练过程中，当验证集的损失不再下降时，停止训练
